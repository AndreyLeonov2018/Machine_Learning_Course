# Центральная предельная теорема своими руками

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

Для решения данной задачи возьмем распределение half-logistic.

```python
from scipy.stats import halflogistic 

# вычисляем параметры исходного определения
EX = math.log(4)
DX = (math.pi**2)/3 - (math.log(4))**2

# генерируем выборку объемом 1000
sample = halflogistic.rvs(size=1000) 

# строим гистрограмму
plt.hist(sample, density=True, bins=10, label='histogram')

# на этом же графике построим теоретическую плотность распределения 
x = np.linspace(0,10,1000)# генерируем массив точек
pdf = halflogistic.pdf(x) # вычисляем значение плотности распределения
plt.plot(x, pdf, label='theoretical pdf', alpha=1)
plt.legend()
plt.ylabel('fraction of samples / f(x)')
plt.xlabel('$x$')
```

## Объем выборки равен 5

```python
# вычисляем  и записываем в массив значения выборочных средних 
i = 1
n = 5
average_list = []
while i <= 1000:
    sample = halflogistic.rvs(size=n)
    average = sum(sample)/n
    average_list.append(average)
    i = i + 1

# вычисляем параметры нормального распределения 
EX_average = EX
DX_average = DX/n
SIGMA_average = math.sqrt(DX_average)

# строим теоретическую плотность нормального распределения
norm_average = sts.norm(EX_average, SIGMA_average)
y = np.linspace(-1,5,100)
pdf = norm_average.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.8)

# строим гистрограмму выборочных средних
plt.hist(average_list, bins = 30, density=True, label='histogram')

plt.legend()
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```
